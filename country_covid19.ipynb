```python
df.skew()    #cheaking skewness of data
```

```python
for col in df.skew().index:
    if col in df.describe().columns:
        if df[col].skew()>0.55:
            df[col]=np.sqrt(df[col])     #treating skewness
        if df[col].skew()<-0.55:
            df[col]=np.cbrt(df[col])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)    #scaling
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression     #building model
```

```python
lr=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =.20,random_state=42)
```

```python
print('x_train.shape =',x_train.shape,'\t\t','x_test.shape =',x_test.shape)
print('y_train.shape =',y_train.shape,'\t\t','y_test.shape =',y_test.shape)
```

```python
df
```

```python
df.info()
```

```python
df
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print('error')

print("mean absolute error =",mean_absolute_error(y_test,pred))
print("mean squared error =",mean_squared_error(y_test,pred))

print("root mean squared error =",np.sqrt(mean_squared_error(y_test,pred)))

print("r2_score =",r2_score(y_test,pred))
```
